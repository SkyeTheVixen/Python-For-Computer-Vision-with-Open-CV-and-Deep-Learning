## Set Up

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
def display_image(img):
    fig = plt.figure(figsize=(12, 10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')

def load_image(path):
    img = cv2.imread(path)
    return img

def fix_image(img):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return img
```

## Sample Images

```python
dark_horse = load_image('DATA/horse.jpg') #BGR For OpenCV
show_horse = fix_image(dark_horse) #Converted for Show

rainbow = load_image('DATA/rainbow.jpg') #BGR For OpenCV
show_rainbow = fix_image(rainbow) #Converted for Show

blue_bricks = load_image('DATA/bricks.jpg') #BGR For OpenCV
show_bricks = fix_image(blue_bricks) #Converted for Show
```

```python
hist_values = cv2.calcHist([dark_horse], channels=[0], mask=None, histSize=[256], ranges=[0, 256])
#cv2.calcHist([image], [channels], mask, [histSize], [ranges])
plt.plot(hist_values)
```

```python
hist_values = cv2.calcHist([rainbow], channels=[1], mask=None, histSize=[256], ranges=[0, 256])
#cv2.calcHist([image], [channels], mask, [histSize], [ranges])
plt.plot(hist_values)
```

```python
hist_values = cv2.calcHist([blue_bricks], channels=[2], mask=None, histSize=[256], ranges=[0, 256])
#cv2.calcHist([image], [channels], mask, [histSize], [ranges])
plt.plot(hist_values)
```

```python
color = ('b', 'g', 'r')

for i, col in enumerate(color):
    histr = cv2.calcHist([blue_bricks], [i], None, [256], [0, 256])
    plt.plot(histr, color=col)
    plt.xlim([0, 256])

plt.title("HISTOGRAM FOR BLUE BRICKS")
plt.xlabel("Value")
```

```python
color = ('b', 'g', 'r')

for i, col in enumerate(color):
    histr = cv2.calcHist([dark_horse], [i], None, [256], [0, 256])
    plt.plot(histr, color=col)
    plt.xlim([0, 30])
    plt.ylim([0, 700000])

plt.title("HISTOGRAM FOR HORSE")
plt.xlabel("Value")
```
