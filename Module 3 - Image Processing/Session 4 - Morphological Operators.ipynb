# Morphological Operators
---

## Set Up

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
def load_image():
    img = np.zeros((600,600))
    Font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(img, "ABCDE", (50, 300), Font, 5, (255, 255, 255), 20)
    #cv2.putText(image, "text", (origin), fontface, fontscale, (color), thickness)
    return img

def display_image(img):
    fig = plt.figure(figsize=(12, 10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
img = load_image()
display_image(img)
```

## Erosion


Erode away the foreground from the background

```python
kernel = np.ones((5, 5), dtype=np.uint8)
```

```python
result = cv2.erode(img, kernel, iterations=5)
# cv2.erode(image, kernel, iterations)
display_image(result)
```

## Opening

```python
img = load_image()
white_noise = np.random.randint(low=0, high=2, size=(600,600))
# np.random.randint(low, high, (size))
display_image(white_noise)
```

```python
white_noise = white_noise * 255
noise_img = white_noise + img
```

```python
display_image(noise_img)
```

```python
opening = cv2.morphologyEx(noise_img, cv2.MORPH_OPEN, kernel)
display_image(opening)
```

## Closing Noise

```python
img = load_image()
black_noise = np.random.randint(0, 2, (600, 600))
black_noise = black_noise * -255
black_noise_img = img + black_noise
black_noise_img[black_noise_img == -255] = 0 # bring - to 0
display_image(black_noise_img)
```

```python
closing = cv2.morphologyEx(black_noise_img, cv2.MORPH_CLOSE, kernel)
display_image(closing)
```

## Morphological Gradient

```python
img = load_image()
display_image(img)
```

```python
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)
display_image(gradient)
```
